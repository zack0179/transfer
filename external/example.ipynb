# Gallery of collinear distributions

```python
import sys,os
import numpy as np
import pylab as py
%matplotlib inline
```

## LSS

```python
from LSSLIB.LSS import LSS
```

```python
conf={'path2LSS':'./LSSLIB'}
lss=LSS(conf)
```

```python
ncols,nrows=1,1
py.figure(figsize=(ncols*7,nrows*5))
ax=py.subplot(nrows,ncols,1)
Q2=10.0
X=np.linspace(0.01,0.99)
ppdf=np.transpose([lss.get_f(x,10) for x in X])
fmap={'g':0,'u':1,'ub':2,'d':3,'db':4,'s':5,'sb':6}
ax.plot(X,X*ppdf[fmap['u']],label='u')
ax.plot(X,X*ppdf[fmap['d']],label='d')
ax.plot(X,X*ppdf[fmap['ub']],label='ub')
ax.plot(X,X*ppdf[fmap['db']],label='du')
ax.plot(X,X*ppdf[fmap['s']],label='s')
ax.plot(X,X*ppdf[fmap['sb']],label='sb')
ax.legend()
ax.set_ylabel('x*ppdf',size=20)
ax.set_xlabel('x',size=20)
```
