```python
from numpy import genfromtxt as gft # Import package
from pandas import DataFrame as df # Import package
import pandas as pd # Import Pandas
```

Read and save the data in excel

```python
L=open('pos_had_crosssec_20x250_new.txt').readlines()     # read each line
L=[l.strip() for l in L]                # remove white space between values in each line
H=L[0].split()    # save the second line as H and delete the hashtag 
dat = gft('pos_had_crosssec_20x250_new.txt')              # Read file from working dir as numpi array
data_raw = df(dat, columns = H)             # save data as a panda data frame with headers
```

```python
# clean the data
data_cleaned = data_raw.query('x != -999')
```

```python
# Charlotte prints three bins in one line, why is it so???
# My solution:
data_new={}
data_new['x']=data_cleaned.x
data_new['Q2']=data_cleaned.Q2
data_new['z']=data_cleaned.z
data_new['y']=data_cleaned.y
data_new['pT']=data_cleaned.pt
data_new['value']=data_cleaned['#crosssection']
data_new['stat_u']=data_cleaned.uncertainty
data_new=pd.DataFrame(data_new)



 


#concatenate the data bins...
eic_data = {}
eic_data = data_new

# add rlative error
eic_data['relative'] = eic_data['stat_u']/eic_data['value']
eic_data['target']=['proton' for i in range(len(eic_data))]
eic_data['hadron']=['pi+' for i in range(len(eic_data))]
eic_data['col']=['EIC' for i in range(len(eic_data))]
eic_data['obs']=['M_EIC' for i in range(len(eic_data))]
eic_data['Ebeam']=[20.0 for i in range(len(eic_data))]
eic_data['Etarget']=[250.0 for i in range(len(eic_data))]




writer=pd.ExcelWriter('9000.xlsx')
eic_data.to_excel(writer,'sheet1')
writer.save() 
```
